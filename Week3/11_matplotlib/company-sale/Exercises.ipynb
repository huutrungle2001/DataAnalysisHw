# Visualizing Company Sale data

### Step 1. Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
```

### Step 2. Import the dataset from this [address](https://github.com/thieu1995/csv-files/blob/main/data/matplotlib/company_sales_data.csv).

```python
df = pd.read_csv("../../Datasets/company_sales_data.csv")
df
```

### Exercise 1: Read Total profit of all months and show it using a line plot

![img](img/01.png)

```python
plt.figure(figsize=(10, 7))
plt.plot(df['month_number'], df['total_profit'], label='Company profit per month')
plt.xlabel('Month Number')
plt.ylabel('Profit in dollars')
plt.xticks(df['month_number'])
plt.title('Company profit per month')
plt.yticks([100000, 200000, 300000, 400000, 500000])
plt.legend(loc='lower right')
plt.show()
```

### Exercise 2: Get total profit of all months and show line plot with the following Style properties

![img](img/02.png)

```python
plt.figure(figsize=(10, 7))
plt.plot(df['month_number'], df['total_profit'],
         label='Company profit per month',
         color='r',
         marker='o',
         markerfacecolor='k',
         linestyle='--',
         linewidth=3)
plt.xlabel('Month Number')
plt.ylabel('Profit data of last year')
plt.xticks(df['month_number'])
plt.title('Company profit per month')
plt.yticks([100000, 200000, 300000, 400000, 500000])
plt.legend(loc='lower right')
plt.show()
```

### Exercise 3: Read all product sales data and show it  using a multiline plot

![img](img/03.png)

```python
plt.figure(figsize=(10, 7))
plt.plot(df['month_number'], df['facecream'], label='Face cream Sales Data', marker='o', linewidth=3)
plt.plot(df['month_number'], df['facewash'], label='Face Wash Sales Data', marker='o', linewidth=3)
plt.plot(df['month_number'], df['toothpaste'], label='ToothPaste Sales Data', marker='o', linewidth=3)
plt.plot(df['month_number'], df['bathingsoap'], label='Bathing Soap Sales Data', marker='o', linewidth=3)
plt.plot(df['month_number'], df['shampoo'], label='Shampoo Sales Data', marker='o', linewidth=3)
plt.plot(df['month_number'], df['moisturizer'], label='Moisturizer Sales Data', marker='o', linewidth=3)
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.xticks(df['month_number'])
plt.title('Sales data')
plt.legend(loc='upper left')
plt.show()
```

### Exercise 4: Read toothpaste sales data of each month and show it using a scatter plot
![img](img/04.png)

```python
# Exercise 4: Read toothpaste sales data of each month and show it using a scatter plot
plt.figure(figsize=(10, 7))
plt.scatter(df['month_number'], df['toothpaste'], label='Tooth paste Sales data')
plt.xlabel('Month Number')
plt.ylabel('Number of units sold')
plt.xticks(df['month_number'])
plt.title('Tooth paste Sales data')
plt.grid(True, linewidth=1, linestyle='--')
plt.legend(loc='upper left')
plt.show()
```

### Exercise 5: Read face cream and facewash product sales data and show it using the bar chart

![img](img/05.png)

```python
# Exercise 5: Read face cream and facewash product sales data and show it using the bar chart

plt.figure(figsize=(10, 7))
plt.bar([a - 0.25 for a in df['month_number']], df['facecream'], width=0.25, label='Face Cream sales data', align='edge')
plt.bar([a + 0.25 for a in df['month_number']], df['facewash'], width=-0.25, label='Face Wash sales data', align='edge')
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.xticks(df['month_number'])
plt.title('Facewash and Facecream sales data')
plt.legend(loc='upper left')
plt.grid(True, linewidth=1, linestyle='--')
plt.show()
```

### Exercise 7: Read the total profit of each month and show it using the histogram to see the most common profit ranges

![img](img/07.png)

```python
# Exercise 7: Read the total profit of each month and show it using the histogram to see the most common profit ranges
plt.figure(figsize=(10, 7))
profit_range = [150000, 200000, 250000, 300000, 350000, 400000, 450000, 500000]
plt.hist(df['total_profit'], profit_range, label='Profit data')
plt.xlabel('Profit range in dollar')
plt.ylabel('Actual Profit in dollar')
plt.title('Profit data')
plt.legend(loc='upper left')
plt.xticks(profit_range)
plt.show()
```

### Exercise 8: Calculate total sale data for last year for each product and show it using a Pie chart

![img](img/08.png)

```python
# Exercise 8: Calculate total sale data for last year for each product and show it using a Pie chart
plt.figure(figsize=(10, 7))
sales_data = [df['facecream'].sum(), df['facewash'].sum(), df['toothpaste'].sum(), df['bathingsoap'].sum(),
              df['shampoo'].sum(), df['moisturizer'].sum()]
labels = ['Face cream', 'Face wash', 'Tooth paste', 'Bathing soap', 'Shampoo', 'Moisturizer']
plt.pie(sales_data, labels=labels, autopct='%1.1f%%')
plt.title('Sales data')
plt.legend(loc='upper left')
plt.show()
```

### Exercise 9: Read Bathing soap facewash of all months and display it using the Subplot

![img](img/09.png)

```python
# Exercise 9: Read Bathing soap facewash of all months and display it using the Subplot

plt.figure(figsize=(10, 7))
plt.subplot(2, 1, 1)
plt.plot(df['month_number'], df['bathingsoap'], label='Bathing soap sales data', marker='o', linewidth=3, color='black')
plt.ylabel('Sales units in number')
plt.title('Sales data')
plt.legend(loc='upper left')
plt.xticks(df['month_number'])

plt.subplot(2, 1, 2)
plt.plot(df['month_number'], df['facewash'], label='Face Wash sales data', marker='o', linewidth=3, color='red')
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.title('Sales data of a facewash')
plt.legend(loc='upper left')
plt.xticks(df['month_number'])
plt.show()
```

### Exercise 10: Read all product sales data and show it using the stack plot

![img](img/10.png)

```python
# Exercise 10: Read all product sales data and show it using the stack plot
plt.figure(figsize=(10, 7))
plt.stackplot(df['month_number'], df['facecream'], df['facewash'], df['toothpaste'], df['bathingsoap'], df['shampoo'],
              df['moisturizer'], labels=['Face cream', 'Face wash', 'Tooth paste', 'Bathing soap', 'Shampoo',
                                         'Moisturizer'])
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.title('All product sales data using stack plot')
plt.legend(loc='upper left')
plt.show()
```
