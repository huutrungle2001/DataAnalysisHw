- `hours-per-week`: continuous.

### 7. (3đ) Tạo pie chart thể hiện phần trăm của cột race và sex (sẽ có 5*2=10 categories tổng thể trong 1 hình vẽ). Nhận xét xem nhóm nào chiếm tỉ lệ nhiều nhất.

```python
race_sex_counts = df.groupby(['race', 'sex']).size()
categories = [f"{race} {sex}" for race, sex in race_sex_counts.index]
percentages = race_sex_counts.values
print(race_sex_counts)
print(categories)
print(percentages)
```

```python
# Create pie chart without labels but with a legend
plt.figure(figsize=(10, 10))
wedges, _, autotexts = plt.pie(
    percentages, labels=[None] * len(categories), autopct='%1.1f%%', startangle=140,
    colors=plt.cm.Paired.colors, pctdistance=0.85, textprops={'fontsize': 10}
)

# Add legend
plt.legend(wedges, categories, title="Race & Sex",
           loc="center left", bbox_to_anchor=(1, 0.5))

plt.title("Distribution of Race and Sex in the Adult Dataset")
plt.show()
```

### 20. (6đ) Tỉ lệ người married hay single men (marital-status feature) chiếm nhiều hơn trong số những người kiếm được >50k? Những người có tình trạng hôn nhân (married) có marital-status feature bắt đầu với Married (Married-civ-spouse, Married-spouse-absent or Married-AF-spouse), còn lại được coi là single men.

```python
married_statuses = ["Married-civ-spouse",
                    "Married-spouse-absent", "Married-AF-spouse"]

high_earners = df[(df["salary"] == ">50K") & (df["sex"] == "Male")]
married_count = high_earners["marital_status"].isin(married_statuses).sum()
single_count = len(high_earners) - married_count

married_count, single_count
```

### 21. (6đ) Số giờ tối đa một người làm việc mỗi tuần (hours-per-week feature) là bao nhiêu? Có bao nhiêu người làm việc với số giờ như vậy và tỷ lệ những người kiếm được nhiều (>50K) trong số đó là bao nhiêu?

```python
max_hours = df["hours_per_week"].max()
max_hours
```

```python
max_hours_count = (df["hours_per_week"] == max_hours).sum()
max_hours_count
```

```python
high_earners_count = ((df["hours_per_week"] == max_hours)
                      & (df["salary"] == ">50K")).sum()
high_earners_percentage = (high_earners_count / max_hours_count) * 100

high_earners_percentage
```
